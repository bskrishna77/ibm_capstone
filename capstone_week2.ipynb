Logistic Regression

Preprocessing the data

```python
X = preprocessing.StandardScaler().fit(bank).transform(bank.astype(float))
X[0:5]
```

```python
from sklearn.model_selection import train_test_split
```

Diving the dataset into training and testing

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

Logistic Regression model building

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(C=0.01, solver='liblinear').fit(X_train,y_train)
LR
```

Predict on test data

```python
yhat = LR.predict(X_test)
yhat
```

Getting the probabilities

```python
yhat_prob = LR.predict_proba(X_test)
yhat_prob
```

Evaluation

```python
print("Logistic Regression's Jaccard similarity: ", jaccard_similarity_score(y_test, yhat))
print("Logistic Regression's Accuracy: ", metrics.accuracy_score(y_test, yhat))
print ("Logistic Regression's F1 Score: ",classification_report(y_test, yhat))
```
